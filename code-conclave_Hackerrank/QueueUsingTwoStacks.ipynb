```python
class Queue:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []
        
#     1. Enqueue
#     2. Dequeue
#     3. Peek

    def enqueue(self, inp):
        self.stack1.append(inp)
        
    def dequeue(self):
        # Check if stack2 is empty, as it will hold the first element on the top.
        # If false pop it
        # If true while stack1 is not empty, pop stack1 elements to stack2
        # then pop stack2
        if not self.stack2:
            if not self.stack1:
                raise IndexError
            while self.stack1:
                self.stack2.append(self.stack1.pop())
        return self.stack2.pop()
    
    def peek(self):
        # Same concept as dequeue, just that we need to show the -1 position item in stack2
        if not self.stack2:
            if not self.stack1:
                raise IndexError
            while self.stack1:
                self.stack2.append(self.stack1.pop())
        return self.stack2[-1]
    
obj = Queue()

q = int(input())

for _ in range(q):
    choice = input().split()
    if choice[0] == "1":
        obj.enqueue(int(choice[1]))
    elif choice[0] == "2":
        obj.dequeue()
    elif choice[0] == "3":
        print(obj.peek())
    else:
        print("Wrong Choice")
```
