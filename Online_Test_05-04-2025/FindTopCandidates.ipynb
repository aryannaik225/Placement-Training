# Find the top cadidates from the given input


```python
data = input().split()

n = int(data[0])
# print(n)

pairs = []

# i = len(data)-1
# while i>0:
#   score = int(data[i-1])
#   name = data[i]
#   pairs.append((name, score))
#   i-=2

# pairs.sort(key=lambda x: x[1], reverse=True)

# for i in range(n):
#   print(f'{pairs[i][0]}: {pairs[i][1]}', end = ' ')

i = 1
while i < len(data)-1:
  name = data[i+1]
  score = int(data[i])
  pairs.append((name, score))
  i+=2
  
pairs.sort(key=lambda x: x[1], reverse=True)

# print(pairs)

for i in range(n):
  print(f"{pairs[i][0]}: {pairs[i][1]}", end=' ')
```
