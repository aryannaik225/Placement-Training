# Check Armstrong in a List


```python
def isArmstrong(num):
  n = num
  mylist = list(str(num))
  sum = 0
  for i in mylist:
    sum += int(i)**len(mylist)
    
  if sum == n:
    return True
  else:
    return False
  


# num = str(153)
# mylist = list(num)
# print(mylist)
  
  
# mylist = list(map(int, input.split(', ')))
# print(mylist)

inp = input()
mylist = [int(x.strip()) for x in inp.split(',')]
# print(mylist)
res = []
for i in mylist:
  if isArmstrong(i):
    res.append(i)

if res:
  for i in res:
    print(i, end=' ')
else:
  print(None)
```
