# Collatz Sequence and Maximum Calculation





```python
def Collatz(num):
  res = []
  while num!=1:
    res.append(num)
    if num%2 == 0:
      num //= 2
    else:
      num = (3*num)+1
  
  res.append(1)
  return res
  

try:
  num = int(input())
  print(Collatz(num))
  # Task 1 done here
  
  # mx_len = 0
  # mx_len_numb = 0
  # mx_num = 0
  # mx_num_pos = -1
  
  # for i in range(1, num+1):
  #   res = Collatz(i)
  #   mx_leng = max(mx_len, len(res))
  #   if mx_leng!=mx_len:
  #     mx_len = mx_leng
  #     mx_len_numb = i
  #   # Task 2 done
    
  #   if mx_num < max(res):
  #     for j in range(res):
  #       if res[j] == max(res):
  #         mx_num = res[j]
  #         mx_num_pos = j
  
  
  max_len = 0
  max_k = 0
  for i in range(1, num+1):
    seq = Collatz(i)
    if len(seq) > max_len:
      max_len = len(seq)
      max_k = i
  print(f"{max_len} {max_k}")
  
  
  
  max_num = 0
  max_num_k = 0
  for i in range(1, num+1):
    seq = Collatz(i)
    seq_max = max(seq)
    if seq_max > max_num:
      max_num = seq_max
      max_num_k = i
  
  print(f"{max_num} {max_num_k}")
  

  # print(f"{mx_len} {mx_len_numb}")
  # print(f"{mx_num} {mx_num_pos}")
  
except:
  print("Error!")
```
