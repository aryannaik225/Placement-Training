### Implementing Stack using Queue

````python
class MyStack(object):

    def __init__(self):
        self.queue1 = []
        self.queue2 = []
        

    def push(self, x):
        """
        :type x: int
        :rtype: None
        """
        self.queue1.append(x)
        

    def pop(self):
        """
        :rtype: int
        """
        if len(self.queue1)<1:
            return
        else:
            while len(self.queue1)!=1:
                self.queue2.append(self.queue1.pop(0))
            res = self.queue1.pop(0)
            while len(self.queue2)!=0:
                self.queue1.append(self.queue2.pop(0))
            return res
        

    def top(self):
        """
        :rtype: int
        """
        if len(self.queue1)<1:
            return
        else:
            while len(self.queue1)!=1:
                self.queue2.append(self.queue1.pop(0))
            res = self.queue1.pop(0)
            self.queue2.append(res)
            while len(self.queue2)!=0:
                self.queue1.append(self.queue2.pop(0))
            return res


    def empty(self):
        """
        :rtype: bool
        """
        if len(self.queue1)<1:
            return True
        else:
            return False
        


# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
````

### Implementing Queue using Stack

````python
class MyQueue(object):

    def __init__(self):
        self.stack1 = []
        self.stack2 = []
        

    def push(self, x):
        """
        :type x: int
        :rtype: None
        """
        if self.stack2:
            while self.stack2:
                self.stack1.append(self.stack2.pop())
        self.stack1.append(x)
        
        

    def pop(self):
        """
        :rtype: int
        """
        if self.empty():
            return
        else:
            while len(self.stack1) != 0:
                self.stack2.append(self.stack1.pop())
            res = self.stack2.pop()
            while len(self.stack2) != 0:
                self.stack1.append(self.stack2.pop())
            return res

        

    def peek(self):
        """
        :rtype: int
        """
        if len(self.stack1)==0:
            return
        else:
            while len(self.stack1) != 0:
                self.stack2.append(self.stack1.pop())
            res = self.stack2[-1]
            while len(self.stack2) != 0:
                self.stack1.append(self.stack2.pop())
            return res

        

    def empty(self):
        """
        :rtype: bool
        """
        if len(self.stack1)==0:
            return True
        else:
            return False
        


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
````

### 🌲 Depth-First Search (DFS) Traversal Orders

#### 🔹 Inorder Traversal (L ➝ R ➝ N)
- **Order**: Left ➝ Root ➝ Right  
- **Used in**: Binary Search Trees (returns sorted order)

#### 🔹 Preorder Traversal (N ➝ L ➝ R)
- **Order**: Root ➝ Left ➝ Right  
- **Used for**: Cloning trees, prefix expressions

#### 🔹 Postorder Traversal (L ➝ R ➝ N)
- **Order**: Left ➝ Right ➝ Root  
- **Used for**: Deleting trees, postfix expressions
